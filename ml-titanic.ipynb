```python
# import library
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.ensemble import RandomForestClassifier
```

```python
# train data
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
train_data.head()
```

```python
# test data
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')
test_data.head()
```

```python
# women survised 
women = train_data.loc[train_data.Sex == 'female']['Survived']
women_survived = sum(women)
women_survived
```

```python
# women survised 
women = train_data [train_data.Sex == 'female'].Survived
women_survived = sum(women) / len(women)
print ('Percentage of women Survived : ',women_survived)
```

```python
# men survised 
men = train_data [train_data.Sex == 'male'].Survived
men_survived = sum(men) / len(men)
print ('Percentage of men Survived : ',men_survived)
```

```python
# target of model
target = train_data ['Survived']
features = ['Pclass', 'Sex', 'SibSp', 'Parch']

# dummey data for model
X_train = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, target)
predictions = model.predict(X_test)
predictions
```

```python
Survived = pd.DataFrame({'PassengerId':test_data.PassengerId, 'Survived':predictions})
Survived.to_csv('titanic_submission_2024_01_23.csv', index=False)
Survived
```
