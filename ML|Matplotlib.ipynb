```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
print(y)
print(z)
```

```python
# sine wave
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine wave')
plt.title('sine wave')
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cosine wave')
plt.title('cosine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,30)
y = x**2
plt.plot(x,y, 'r.')
plt.show()
```

```python
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'g--')
plt.show()
```

### Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ['EN', 'BN', 'IN','FR']
people = [102,33,22,91]
ax.bar(lang,people)
plt.xlabel ('Languages')
plt.ylabel ('Number of People')
plt.show()
```

### Pia Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ['EN', 'BN', 'IN','FR']
people = [102,33,22,91]
ax.pie(people, labels = lang, autopct = '%1.1f%%')
plt.show()
```

### Scatter Plot

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z, cmap='Blues')
plt.show()
```
